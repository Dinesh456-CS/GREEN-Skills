```python
def calculate_carbon_footprint(energy_consumption,emmission_factor):
    return energy_consumption*emmission_factor
```

```python
energy_consumption=1000
emmission_factor=0.475
carbon_footprint=calculate_carbon_footprint(energy_consumption,emmission_factor)
print(f"carbon footprint: {carbon_footprint} Kg CO2")
```

```python
import pandas as pd
import numpy as np
data={
    "Energy Source":["Solar", "Wind", "Hydropower", "Geothermal", "Biomass","Nuclear"],
    "Energy Consumption (MWh)":[1200,np.nan,2900,np.nan,2500,3200],
    "Cost (Million $)":[200,400,np.nan,150,250,np.nan]
}
energy_df=pd.DataFrame(data)
print("orginal energy data with missing values:")
print(energy_df)
```

```python
cleaned_df=energy_df.dropna()
print("\n data after removing rows with missing values ")
print(cleaned_df)
```

```python
forward_filled_df=energy_df.fillna(method="ffill")
print("\n data after forward filling")
print(forward_filled_df)
```

```python
energy_df["Energy Consumption (MWh)"].fillna(energy_df["Energy Consumption (MWh)"].mean(),inplace=True)
energy_df["Cost (Million $)"].fillna(energy_df["Cost (Million $)"].mean(),inplace=True)
print("\n Data after imputing missing values with mean")
print(energy_df)
```

```python
energy_df["Missing Consumption"]=energy_df["Energy Consumption (MWh)"].isna().astype(int)
print("\n data with missing values flagged")
print(energy_df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
energy_df[["Energy Consumption (MWh)","Cost (Million $)"]]=scaler.fit_transform(
           energy_df[["Energy Consumption (MWh)","Cost (Million $)"]]
)
print("\n Data after normalization (Min Max Scalling):")
print(energy_df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
energy_df[["Energy Consumption (MWh)","Cost (Million $)"]]=scaler.fit_transform(
           energy_df[["Energy Consumption (MWh)","Cost (Million $)"]]
)
print("\n Data after Standardization (z-score Scalling):")
print(energy_df)
```

```python
# one-hot Encode
energy_encoded_df=pd.get_dummies(energy_df,columns=["Energy Source"])
print("\n data after one-hot encoding categorical variables:")
print(energy_encoded_df)
```

```python
energy_encoded_df["Consumption per $Million"]=energy_encoded_df["Energy Consumption (MWh)"]/energy_encoded_df["Cost (Million $)"]
print("\n Data with new freature (consumption per $Million)")
print(energy_encoded_df)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Solar":[1200,1500,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass":[2500,2700,2400]
}
df=pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('correlation between energy sources')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Distribution of energy consumption by source')
plt.show()
```

```python
sns.violinplot(data=df)
plt.title('violin plotn of energy consumption by source')
plt.show()
```

```python
energy_values=[100,200,300,400,500]
carbon_emission=[10,20,30,40,50]
df_reg=pd.DataFrame({
    'Energy Consumption':energy_values,
    'Energy Emission':carbon_emission
})
sns.regplot(x='Energy Consumption',y='Energy Emission',data=df_reg)
plt.title('Energy Consumption vs carbon Emission')
plt.xlabel('Energy consumption')
plt.ylabel('Carbon Emission')
plt.show()
```
